#### USING THE GIVEN DATA FILE 'data.csv'
---

```python
# assigning datase to the information in the data.csv file.
dataframe = pd.read_csv('data.csv')
```

#### ANALYSING AND REMOVING ALL NULL VALUES AND DUPLICATES
---

```python
# Replacing the spaces between words in the columns with underscores '_'
dataframe.columns = dataframe.columns.str.replace(' ','_')
```

```python
dataframe
```

```python
#Checking the data information
dataframe.info()
```

The isnull() method is used to check whether a value in a pandas DataFrame or series is missing or not. This method returns a boolean mask indicating where the values are null(True) or non null(false)

```python
# Checking for the total number of null values
dataframe.isnull().sum()
```

```python
#Finding the mean values of each column
meanDATA = dataframe.mean()
```

```python
# This variable represents the mean of each of the dataframe
meanDATA
```

#### REPLACING ALL THE NULL VALUES WITH THE MEAN OF EACH COLUMN AND REMOVING ALL DUPLICATES
---

```python
dataframe = dataframe.fillna(meanDATA).drop_duplicates()
```

```python
meanFA = dataframe.fixed_acidity.mean()
meanVA = dataframe.volatile_acidity.mean()
meanCA = dataframe.citric_acid.mean()
meanRS = dataframe.residual_sugar.mean()
meanC = dataframe.chlorides.mean()
meanFSD = dataframe.free_sulfur_dioxide.mean()
meanTSD = dataframe.total_sulfur_dioxide.mean()
meanD = dataframe.density.mean()
meanPH = dataframe.pH.mean()
meanS = dataframe.sulphates.mean()
meanA = dataframe.alcohol.mean()
meanQ = dataframe.quality.mean()
```

#### GRAPHING THE DATA
___

```python
# data for pie chart
# total columns excluding ID is 12
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple',
          'pink', 'brown', 'gray', 'black', 'violet', 'beige']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5, 0, 0.5)
meanValues = [meanFA, meanVA, meanCA, meanRS, meanC,
              meanFSD, meanTSD, meanD, meanPH, meanS, meanA, meanQ]
meanTopics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
xtitle = 'Column names'
ytitle = 'Mean Values'
legendLabels = ['fixed_acidity (FA)',
                'volatile_acidity (VA)',
                'citric_acid (CA)',
                'residual_sugar (RS)',
                'chlorides (C)',
                'free_sulfur_dioxide (FSD)',
                'total_sulfur_dioxide (TSD)',
                'density',
                'pH',
                'sulphates',
                'alcohol',
                'quality']
```

```python
labels = meanTopics
sizes = meanValues
```

#### COMPARING THE MEAN VALUES OF THE DATA USING A PIE CHART
___

```python
# plotting the pie chart
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colors, labeldistance=1.2)
plt.axis('equal')
plt.legend(title='index', labels=legendLabels, loc='upper right',)
plt.title('Mean Values for each of the columns in a pie chart')
plt.figure(figsize=(15, 18))
plt.show()
```

#### COMPARING THE MEAN VALUES OF THE DATA USING A BAR GRAPH
___

```python
# plotting a bar graph

fig = plt.figure(figsize = (10, 5))

plt.bar(labels, sizes, width = 0.5, color = colors)
plt.xlabel(xtitle)
plt.ylabel(ytitle)
plt.title('Comparing mean of columns in a bar chart')
plt.show()
```
